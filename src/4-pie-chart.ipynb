### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
```

```python
import sys
import subprocess

# Check and install matplotlib if not available
try:
    import matplotlib.pyplot as plt
except ImportError:
    print("matplotlib not found. Installing now...")
    subprocess.check_call([sys.executable, "-m", "pip", "install", "matplotlib"])
    import matplotlib.pyplot as plt

# Task: Create a pie chart with categories and values

# Define labels and sizes arrays
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 35, 20, 20]

# Check if labels and sizes arrays are the same length
if len(labels) != len(sizes):
    raise ValueError("Labels and sizes arrays must be of the same length before plotting.")

# Create pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Add title
plt.title("Pie Chart of Categories")

# Ensure pie is drawn as a circle
plt.axis('equal')

# Display plot
plt.show()
```
