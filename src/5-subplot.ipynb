### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import sys
import subprocess

# Check and install matplotlib if not available
try:
    import matplotlib.pyplot as plt
except ImportError:
    print("matplotlib not found. Installing now...")
    subprocess.check_call([sys.executable, "-m", "pip", "install", "matplotlib"])
    import matplotlib.pyplot as plt

# --- Data for plots ---

# Line plot data (y = x²)
x_line = list(range(0, 11))
y_line = [i**2 for i in x_line]

# Bar plot data
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]

# Scatter plot data
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 1, 8, 7]

# Pie chart data
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 35, 20, 20]

# --- Validation for robustness ---
if len(x_line) != len(y_line):
    raise ValueError("Line plot: x and y arrays must be the same length.")
if len(categories) != len(values):
    raise ValueError("Bar plot: categories and values arrays must be the same length.")
if len(x_scatter) != len(y_scatter):
    raise ValueError("Scatter plot: x and y arrays must be the same length.")
if len(labels) != len(sizes):
    raise ValueError("Pie chart: labels and sizes arrays must be the same length.")

# --- Create 2x2 grid of plots ---
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line plot (top-left)
axs[0, 0].plot(x_line, y_line, marker='o', linestyle='-', color='b', label="y = x²")
axs[0, 0].set_title("Line Plot")
axs[0, 0].set_xlabel("x")
axs[0, 0].set_ylabel("y")
axs[0, 0].grid(True)
axs[0, 0].legend()

# Bar plot (top-right)
axs[0, 1].bar(categories, values, color='skyblue')
axs[0, 1].set_title("Bar Plot")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")
axs[0, 1].grid(axis='y', linestyle='--', alpha=0.7)

# Scatter plot (bottom-left)
axs[1, 0].scatter(x_scatter, y_scatter, color='green', marker='o')
for i in range(len(x_scatter)):
    axs[1, 0].annotate(f'({x_scatter[i]}, {y_scatter[i]})', (x_scatter[i], y_scatter[i]), 
                       textcoords="offset points", xytext=(5,5), ha='center')
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("x")
axs[1, 0].set_ylabel("y")
axs[1, 0].grid(True)

# Pie chart (bottom-right)
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title("Pie Chart")
axs[1, 1].axis('equal')

# Prevent overlap between subplots
plt.tight_layout()

# Show all plots
plt.show()
```
