```python
try:
    import matplotlib.pyplot as plt
except ImportError:
    import sys
    !{sys.executable} -m pip install matplotlib
    import matplotlib.pyplot as plt

import numpy as np

# Data
x = np.arange(1, 6)
y = x * 2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'green', 'blue', 'orange']

pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart (subplot 1)
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
axs[0, 0].set_title('Line Chart')

# Bar chart (subplot 2)
axs[0, 1].bar(categories, values, color=colors)
axs[0, 1].set_title('Bar Chart')

# Scatter plot (subplot 3)
axs[1, 0].scatter(x, y, color='purple')
for i in range(len(x)):
    axs[1, 0].annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='center')
axs[1, 0].set_title('Scatter Plot')

# Pie chart (subplot 4)
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart')

# Adjust layout and show
plt.tight_layout()
plt.show()
```
