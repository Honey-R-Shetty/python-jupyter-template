### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
import sys
import subprocess

# Check and install matplotlib if not available
try:
    import matplotlib.pyplot as plt
except ImportError:
    print("matplotlib not found. Installing now...")
    subprocess.check_call([sys.executable, "-m", "pip", "install", "matplotlib"])
    import matplotlib.pyplot as plt

# Task: Create a bar plot with distinct categories and values

# Define categories and values arrays
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]

# Check if categories and values arrays are the same length
if len(categories) != len(values):
    raise ValueError("Categories and values arrays must be of the same length before plotting.")

# Create bar plot
plt.bar(categories, values, color='skyblue')

# Add title and axis labels
plt.title("Bar Plot of Categories and Values")
plt.xlabel("Categories")
plt.ylabel("Values")

# Add grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display plot
plt.show()
```
