### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
import sys
import subprocess

# Check and install matplotlib if not available
try:
    import matplotlib.pyplot as plt
except ImportError:
    print("matplotlib not found. Installing now...")
    subprocess.check_call([sys.executable, "-m", "pip", "install", "matplotlib"])
    import matplotlib.pyplot as plt

# Task: Create a histogram from a given data array

# Define data array
data = [12, 15, 13, 15, 17, 18, 15, 14, 16, 18, 19, 17, 18, 15, 16]

# Check if data array is empty
if len(data) == 0:
    raise ValueError("Data array is empty. Cannot plot histogram.")

# Create histogram
plt.hist(data, bins=5, color='skyblue', edgecolor='black')

# Add title and axis labels
plt.title("Histogram of Data Values")
plt.xlabel("Value Range")
plt.ylabel("Frequency")

# Add grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display plot
plt.show()
```
