### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import sys
import subprocess

# Check and install matplotlib if not available
try:
    import matplotlib.pyplot as plt
except ImportError:
    print("matplotlib not found. Installing now...")
    subprocess.check_call([sys.executable, "-m", "pip", "install", "matplotlib"])
    import matplotlib.pyplot as plt

# Task: Create a scatter plot with x and y coordinates, with point annotations

# Define x and y arrays
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 8, 7]

# Check if x and y arrays are the same length
if len(x) != len(y):
    raise ValueError("x and y arrays must be of the same length before plotting.")

# Create scatter plot
plt.scatter(x, y, color='green', marker='o')

# Annotate points â€” but limit to max 10 points to avoid clutter (Line 18 enhancement)
max_annotations = 10
for i in range(min(len(x), max_annotations)):
    plt.annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='center')

# Add title and axis labels
plt.title("Scatter Plot with Annotations")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# Add grid
plt.grid(True)

# Display plot
plt.show()
```
